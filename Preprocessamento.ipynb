```python
import pandas as pd
import numpy as np
```

```python
spotify_songs = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-01-21/spotify_songs.csv')
spotify_songs
```

# 1 Erros na formatacao da data

```python
# Busca por erros na formatacao da data

# Tenta converter para datetime, forcando os erros a se tornarem 'NaT'
test_dates = pd.to_datetime(spotify_songs["track_album_release_date"], format="%Y-%m-%d", errors="coerce")

# Encontra as linhas onde a conversao falhou
formatting_errors = spotify_songs[test_dates.isna()]

print(formatting_errors["track_album_release_date"])
```

```python
# Formata a data para que contenha apenas o ano, criando a coluna "release_year"
# Remove a antiga coluna "track_album_release_date"

# 'format="mixed"' tells pandas to figure out if it's "YYYY-MM-DD" or just "YYYY"
spotify_songs["release_year"] = pd.to_datetime(spotify_songs["track_album_release_date"], format="mixed").dt.year

spotify_songs = spotify_songs.drop(columns=["track_album_release_date"])

# Imprime os anos em que as musicas do dataset foram lancados
spotify_songs["release_year"].unique()
```

```python
# Busca por dados nulos no dataset inteiro
spotify_songs.isnull().sum()

# Cria um filtro que procura por valores nulos
null_filter = spotify_songs.isnull().any(axis=1)

# Aplica o filtro para ver as linhas
rows_with_nulls = spotify_songs[null_filter]

print(rows_with_nulls)
```

```python
# Remove qualquer linha que contenha pelo menos um NaN
spotify_songs = spotify_songs.dropna()

# Reseta o indice para que o numero das linhas seja perfeitamente sequencial
spotify_songs = spotify_songs.reset_index(drop=True)

spotify_songs.shape
```

# 2.1 Removendo valores duplicados

```python
# Encontra todas as linhas que compartilham o mesmo track_id
duplicates_only = spotify_songs[spotify_songs.duplicated(subset=["track_id"], keep=False)]

# Ordena eles pelo track_id
duplicates_sorted = duplicates_only.sort_values("track_id")

# Imprime as primeiras 10 linhas
print(duplicates_sorted[["track_name", "playlist_genre", "danceability", "energy"]].head(10))

print(f"Linhas duplicadas: {len(duplicates_only)}")
```

```python
linhas_antes = len(spotify_songs)
print(f"Linhas antes: {linhas_antes}")

# 1. Fazer o One-Hot Encoding focado no subgênero
# Isso vai criar as colunas de subgêneros com 0s e 1s.
subgeneros_ohe = pd.get_dummies(spotify_songs[['track_id', 'playlist_subgenre']], columns=['playlist_subgenre'], dtype=int)

# 2. Agrupar pelo ID da música e pegar o valor máximo
# Como as colunas são 0 ou 1, o .max() funciona como uma porta lógica OR.
# Se a música era do subgênero X na linha 1 e Y na linha 2, ambas as colunas X e Y ficarão com 1.
subgeneros_agrupados = subgeneros_ohe.groupby('track_id').max().reset_index()

# 3. Remover a coluna categórica original e limpar as duplicatas das outras features
# Retiramos 'playlist_subgenre' porque já a transformamos em várias colunas,
# e agora podemos usar com segurança o drop_duplicates para as features de áudio (que não mudam)
spotify_songs_unique = spotify_songs.drop(columns=['playlist_subgenre']).drop_duplicates(subset=['track_id'])

# 4. Juntar as features únicas com os subgêneros agregados
spotify_songs_final = pd.merge(spotify_songs_unique, subgeneros_agrupados, on='track_id', how='left')

linhas_depois = len(spotify_songs_final)
print(f"Linhas depois: {linhas_depois}")
print(f"Total de removidos: {linhas_antes - linhas_depois}")
print(f"Total de músicas resolvidas: {len(duplicates_only) - (linhas_antes - linhas_depois)}")
```

Imprimindo uma música após arrumar

```python
# 1. Configurar o Pandas para não esconder colunas no print (opcional, mas recomendado)
pd.set_option('display.max_columns', None)

# 2. Pegar o primeiro track_id da sua lista de duplicados para usar como cobaia
id_teste = duplicates_sorted['track_id'].iloc[0]

print(f"===========================================================")
print(f"🔍 ANALISANDO A MÚSICA COM ID: {id_teste}")
print(f"===========================================================\n")

# 3. Mostrar as linhas ANTES (no dataset original)
print("[ANTES] Linhas originais (note os subgêneros diferentes):")
antes = spotify_songs[spotify_songs['track_id'] == id_teste]
# Mostrando apenas colunas chave para facilitar a visualização do problema
print(antes[['track_id', 'track_name', 'playlist_genre', 'playlist_subgenre']])
print("\n")

# 4. Mostrar a linha DEPOIS (no dataset final)
print("[DEPOIS] Linha única com os subgêneros agregados:")
depois = spotify_songs_final[spotify_songs_final['track_id'] == id_teste]

# Para provar que a agregação funcionou, vamos listar apenas as colunas de subgênero
# que receberam o valor 1 para esta música específica
colunas_subgenero = [col for col in depois.columns if 'playlist_subgenre_' in col]
subgeneros_ativados = depois[colunas_subgenero].loc[:, depois[colunas_subgenero].eq(1).any()]

# Imprime os metadados básicos e os subgêneros que ficaram com '1'
print(depois[['track_id', 'track_name', 'playlist_genre']])
print("\nSubgêneros ativados (com valor 1) nesta linha:")
print(subgeneros_ativados.to_string(index=False))

# Se você quiser imprimir literalmente a linha inteira com todas as centenas de colunas:
print("\nLinha Completa Transposta:")
print(depois.T)
```

```python
# 1. Definir as colunas que são apenas informativas/identificadoras
colunas_metadados = [
    'track_id', 'track_name', 'track_artist', 'track_popularity',
    'track_album_id', 'track_album_name', 'playlist_name',
    'playlist_id', 'playlist_genre', 'release_year'
]

# Garantir que não vamos tentar remover uma coluna que talvez já não exista no df
colunas_metadados_presentes = [col for col in colunas_metadados if col in spotify_songs_final.columns]

# 2. Criar o DataFrame de Metadados (Guardar para o sistema de recomendação visual)
df_metadados = spotify_songs_final[colunas_metadados_presentes].set_index('track_id')

# 3. Criar o DataFrame de Features (Para o modelo de Machine Learning)
# Removemos os metadados (exceto o track_id, que vamos transformar em índice)
df_features = spotify_songs_final.drop(columns=[col for col in colunas_metadados_presentes if col != 'track_id'])
df_features = df_features.set_index('track_id')

# 4. Verificação final
print(f"Dimensões do df_metadados: {df_metadados.shape}")
print(f"Dimensões do df_features: {df_features.shape}\n")

print("Tipos de dados no df_features (devem ser apenas numéricos: float, int, uint8):")
print(df_features.dtypes.value_counts())
```

Removendo quando features = -1 (If no key was detected, the value is -1.)

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# ---------------------------------------------------------
# PASSO 1: Tratamento de Outliers e Anomalias
# ---------------------------------------------------------

# Capturando a quantidade de linhas antes da limpeza
linhas_antes_limpeza = len(df_features)

print("===========================================================")
print("📊 INICIANDO LIMPEZA DE ANOMALIAS E OUTLIERS")
print("===========================================================")
print(f"Linhas antes da limpeza: {linhas_antes_limpeza}")

# Remover faixas onde a tonalidade não foi detectada (-1)
df_features = df_features[df_features['key'] != -1]

# Filtrar durações extremas (mantendo músicas entre 1 min e 10 min)
min_duration = 60000    # 1 minuto em ms
max_duration = 600000   # 10 minutos em ms
df_features = df_features[(df_features['duration_ms'] >= min_duration) &
                          (df_features['duration_ms'] <= max_duration)]

# Precisamos garantir que o df_metadados acompanhe essas remoções!
df_metadados = df_metadados.loc[df_features.index]

# Capturando a quantidade de linhas depois e calculando a diferença
linhas_depois_limpeza = len(df_features)
total_removidos = linhas_antes_limpeza - linhas_depois_limpeza

print(f"Linhas depois da limpeza: {linhas_depois_limpeza}")
print(f"Total de músicas anômalas/outliers removidas: {total_removidos}\n")
```

One-Hot Encoding do 'key'

```python
# Transforma a coluna 'key' (0 a 11) em 12 colunas binárias independentes
df_features = pd.get_dummies(df_features, columns=['key'], prefix='key', dtype='int8')
```

Escalonamento (Min-Max Scaling)

```python
# Separar colunas que já são binárias (subgêneros, modes e keys)
# das que são contínuas e precisam de normalização
colunas_binarias = [col for col in df_features.columns if 'subgenre' in col or 'key_' in col or col == 'mode']
colunas_continuas = [col for col in df_features.columns if col not in colunas_binarias]

# Instanciar e aplicar o scaler apenas nas colunas contínuas
scaler = MinMaxScaler()
df_features[colunas_continuas] = scaler.fit_transform(df_features[colunas_continuas])
```

Imprimindo

```python
print("===========================================================")
print("✅ PIPELINE DE PRÉ-PROCESSAMENTO FINALIZADO!")
print("===========================================================")
print(f"Dimensões finais da matriz de features: {df_features.shape}")
print("\nValores mínimos após scaling (Amostra):")
print(df_features[colunas_continuas].min().head(3))
print("\nValores máximos após scaling (Amostra):")
print(df_features[colunas_continuas].max().head(3))
```
